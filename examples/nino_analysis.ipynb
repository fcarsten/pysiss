```python
import numpy
import scipy
import matplotlib.pyplot as plt
from scipy.io import netcdf
from cwavelets import Morlet, Paul, Hermitian
from cwavelets.plotting import plot_signal
```

```python
# Open netCDF file
filename = 'nino3.nc'
nino3 = scipy.io.netcdf_file(filename, 'r')
sst_data = nino3.variables['NINO3'].data

# Get time location - NOAA times taken relative to 1871
dt = 1 / 12. # 1 recod per month
time = numpy.arange(len(sst_data))*dt + 1871.0

# Plot signal
figure(figsize=(14,3))
ax = gca()
ax.set_xlim(1871, 2008)
ax.set_title('Nino3 SST dataset')
ax.set_xlabel('Year')
ax.set_ylabel('SST Anomaly')
plot_signal(ax, time, sst_data)
```

```python
wavs = []
for wavelet in [Morlet, Hermitian]:
    wav = wavelet(sst_data.copy(), domain=time.copy(), properties=dict(base=2, fundamental_frequency=6., order=2, scaling='log', scale_range=(1,64), pad=True))
    wav.detrend('cubic')
    wavs.append(wav)
```

```python
map(lambda x: x.spectrogram(figsize=(12,6)), wavs)
```

```python
map(lambda x: x.phase_plot(), wavs)
```
