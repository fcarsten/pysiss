The aim here is to demo a workflow which goes from Level 0 in the Data Maturity Model ('mental' data integration by individual scientists) through Level 1 (data-driven discovery and analysis) to Level 2 (uncertainty quantification). We're going to use my wavelets module to perform domaining based on continuous wavelet transforms using Hermitian wavelets (similar to what is done for object identification in the computer vision field) and then PyMC to perform MCMC-based Bayesian estimation of lithology types within the given boreholes. 

At the end of this workflow we will have a classification schema for the boreholes based on the geochemical data available which assigns observations to lithology types with quantified uncertainty. ALthough the methods used here are pretty crude, at least this will give us something to hang our hats on for the next year or so.

Before munging this data, we'd better say something about what it actually is.

We need to generate a dictionary with some `PropertyType` instances which contain metadata about the properties we're going to define. Here's a few. (_Longer term this needs to be replaced with some clever geo-vocab stuff from Simon Cox et al._)

```python
# Labels and long-form version
long_names = dict(zip(vogue_data.keys(), itertools.repeat(None)))
long_names.update({
    'As_ppm': 'Arsenic',
    'CaO_pct': 'Calcium',
    'Co_ppm': 'Cobalt',
    'Cr_ppm': 'Chromium',
    'Cu_ppm': 'Copper',
    'Fe2O3_pct': 'Ferric Iron (3+)',
    'K2O_pct': 'Potassium',
    'MnO_pct': 'Manganese',
    'Ni_ppm': 'Nickel',
    'Rb_ppm': 'Rubidium',
    'SO3_pct': 'Sulphur',
    'Sr_ppm': 'Strontium',
    'TiO2_pct': 'Titanium',
    'V_ppm': 'Vanadium',
    'Zn_ppm': 'Zinc',
    'Zr_ppm': 'Zirconium',
    'Sample': 'Sample number',
    'Hole_ID': 'Borehole ID',
    'From': 'Depth from',
    'To': 'Depth to',
    'TYPE': 'Measurement method'
})

# Set units
def units(key):
    """ Returns the units for the given key
    """
    if 'ppm' in key:
        return 'ppm'
    elif 'pct' in key and 'O' in key:
        return 'oxide wt. %'
    elif key in ['From', 'To'] or 'mid' in key:
        return 'metres'
    else:
        return None
```

We can then use the dictionaries to generate PropertyType instances, which hold metadata about the properties we're defining in our borehole.

```python
property_types = dict()
for key in vogue_data.keys():
    property_types[key] = borehole_analysis.PropertyType(name=key,
                                                         long_name=long_names[key],
                                                         units=units(key))

# Just to check, print first 5 properties
for ptype in property_types.values()[:5]:
    print ptype
```

Because our borehole implementation expects things to be ordered, we're going to pre-sort the data. Whether this is a good thing to be imposing at this stage I'm not sure - it could get quite nasty if you're trying to add a big dataset...

If we just plot the 'From' and 'To' data we see that things are pretty unordered

```python
figure(figsize=(12, 3))
plot(vogue_data['From'], 'k-', label='From')
plot(vogue_data['To'], 'r-', label='To')
legend(loc='best')
```

We can sort the data using pandas' inbuilt sorting - here we sort first on the 'Hole_ID' attribute, then on the 'From' attribute and then on the 'To' attribute.

```python
vogue_data = vogue_data.sort(('Hole_ID', 'From', 'To')) 
```

Replotting the data we see that things are much better organised - we now have extra  intervals starting at zero and extending to the first sample in the borehole. Although the properties associated with these starting intervals are all NaN, we leave them in for now. We can mask on NaNs later when we have functions which are not NaN-aware.

```python
# Show the sorting
figure(figsize=(12, 3))
plot(vogue_data['From'], 'k-', label='From')
plot(vogue_data['To'], 'r-', label='To')
legend(loc='best')
```

OK, now we can group the data by borehole, generate domains and add these to some Borehole instances. The `groupby` method splits our dataset into categories determined by the supplied key ('Hole_ID' in this case) and returns a Python iterable which we can use to generate each `Borehole` instance:

```python
property_keys = ['As_ppm', 'CaO_pct', 'Co_ppm', 'Cr_ppm', 'Cu_ppm', 'Fe2O3_pct', 'K2O_pct', 'MnO_pct', 'Ni_ppm', 'Rb_ppm', 'SO3_pct', 'Sr_ppm', 'TiO2_pct', 'V_ppm', 'Zn_ppm', 'Zr_ppm']

boreholes = dict()
for name, data in vogue_data.groupby('Hole_ID'):
    # Construct the borehole instance first
    bh = boreholes[name] = borehole_analysis.Borehole(name)

    # Generate an IntervalDomain corresponding to the values in the spreadsheet
    data = data.dropna()
    from_depths = numpy.asarray(data['From'])
    to_depths = numpy.asarray(data['To'])
    domain = bh.add_interval_domain('geochemistry', from_depths, to_depths)
    
    # Add all the geochemistry values as Properties
    for key in property_keys:
        datum = numpy.asarray(data[key].convert_objects(convert_numeric=True))
        domain.add_property(property_types[key], datum)
```

Here we've had to make sure that the data are floats and specified as numpy arrays. While pandas does a good job of automatically identifying the type of the data in our input file, some of the values can't be converted to floats. For example, some of the values are given as `'< 10'`, which has no float representation. When pandas finds a value like this it assumes that the entire column can't be converted and leaves everything as strings. While this is a nice default in general, here we'd prefer to simply mark values which can't be converted as NaN.

This is done with the `convert_objects` method on the pandas `DataFrame`, so

    datum = numpy.asarray(data[key].convert_objects(convert_numeric=True))

forces the data we get back to be numpy arrays of floats which we can then pass to our borehole domain as property data. 

(_Note that this should and could be done automatically with the PropertyType instance - if a PropertyType has a numeric flag set it should force reading everything as floats or NaNs_)

Just to demonstrate how we can get the data back for calculations or visualisation, we can plot the midpoint locations of all the IntervalDomains from all the boreholes to check that everything is ordered and we don't have any strange values

```python
figsize(11, 6)
for bh in boreholes.values():
    domain = bh.interval_domains['geochemistry']
    mid_points = (domain.from_depths + domain.to_depths) / 2.
    plot(mid_points, label=bh.name)
legend(ncol=6, loc=(0, 1.01))
xlabel('Sample number')
ylabel('Depth midpoint (m)')
xlim(-5, 900)
tight_layout()
```

### 2. Identifying gaps and resampling

To do the wavelet FFTs we require an evenly-sampled dataset with no gaps.

We can start by generating a rough plot of some of the signals in an individual borehole

```python
figure(figsize=(10, 6))
for key in property_keys:
    # Get information from the borehole
    domain = bh.interval_domains['geochemistry']
    signal = domain.properties[key].values
    mid_points = (domain.from_depths + domain.to_depths) / 2.
    label = domain.properties[key].property_type.long_name  
    
    # Actually plot data
    plot(mid_points, (signal - signal.mean()) / signal.std(), '.', label=label)

legend(ncol=1, loc=(1.01, 0))
xlabel('Depth midpoint (m)')
ylabel(r'Normalised signal $(x - \bar{x})/ \sigma_{x}$')
ylim(-4, 4)
title('Data for {0}'.format(bhname))
tight_layout()
```

We've got a number of gaps that we'd like to keep track of within the analysis to come. But we need to find this programmatically rather than manually.

Because we have an `IntervalDomain` instance we can just examine the gaps between intervals - if there's a gap we can mark it and move on

```python
intdom = bh.interval_domains['geochemistry']
subdomains, gaps = intdom.split_at_gaps()

print 'Subdomains:', subdomains
print 'Gaps:', gaps
```

The `split_at_gaps` method defines both a `gaps` attribute and a `subdomains` attribute in the `IntervalDomain` instance, given as a list of `(from_depth, to_depth)` tuples (these are set to `None` until this method is run). You can then use these tuples, together with the `get_interval` method, to generate a series of `IntervalDomain`s for each subdomain.

```python
ylim(-0.5, 0.5)
ylabel(r'Normalised As $(x - \bar{x}) / \sigma_{x}$')
xlabel('Midpoint depth (m)')

# Add gap ranges
# Note that we can also query subdom.gaps if we didn't store gaps in the previous cell
for gap in gaps:
    fill_betweenx([-0.25, 0.25], gap[0], gap[1], color='black', alpha=0.1)

# Plot up arsenic data seperated by subdomain
meanval = intdom.properties['As_ppm'].values.mean()
stdval = intdom.properties['As_ppm'].values.std()
for sdom_num, sdom_range in enumerate(subdomains):
    # We need to just get the interval 
    sdom = intdom.get_interval(*sdom_range)
    data = sdom.properties['As_ppm'].values
    line = plot((sdom.from_depths + sdom.to_depths) / 2., (data - meanval) / stdval, '.', 
        label='Subdomain {0}'.format(sdom_num))
    fill_betweenx([-0.25, 0.25], sdom_range[0], sdom_range[1], 
        color=line[0].get_color(), alpha=0.2)
legend(loc='best', ncol=4)
```

If we only have a SamplingDomain, then every sample essentially has a gap around itself, so we need some extra infomation about how big a spacing has to be to qualify as a 'gap'. A useful measure would be a gap size relative to the median sample space (we take the median rather than the mean since the mean may be skewed by potentially large gaps in the data).

Just to demonstrate this method, we'll generate a SamplingDomain and do some gap detection on it using the median sample spacing as our reference spacing, and define gaps as a distance larger than an order of magnitude than this

```python
# Calculate the midpoints of our geochemistry IntervalDomain
intdom = bh.interval_domains['geochemistry']
midpoints = (intdom.from_depths + intdom.to_depths) / 2.

# Add a new SamplingDomain to the borehole
sampdom = bh.add_sampling_domain('geochemistry', midpoints)
for key in property_keys:
    sampdom.add_property(property_types[key], intdom.properties[key].values)

print bh
print '\t--> ', bh.interval_domains.values()
print '\t--> ', bh.sampling_domains.values()

# Find and store gaps
subdomains, gaps = sampdom.split_at_gaps(gap_metric='spacing_median')
```

The `split_at_gaps` method behaves the same for a `SamplingDomain` as an `IntervalDomain`:

```python
ylim(-0.5, 0.5)
ylabel(r'Normalised As $(x - \bar{x}) / \sigma_{x}$')
xlabel('Midpoint depth (m)')

# Add gap ranges
# Note that we can also query subdom.gaps if we didn't store gaps in the previous cell
for gap in gaps:
    fill_betweenx([-0.25, 0.25], gap[0], gap[1], color='black', alpha=0.1)

# Plot up arsenic data seperated by subdomain
meanval = sampdom.properties['As_ppm'].values.mean()
stdval = sampdom.properties['As_ppm'].values.std()
for sdom_num, sdom_range in enumerate(subdomains):
    # We need to just get the interval 
    sdom = sampdom.get_interval(*sdom_range)
    data = sdom.properties['As_ppm'].values
    line = plot(sdom.depths, (data - meanval) / stdval, '.', 
        label='Subdomain {0}'.format(sdom_num))
    fill_betweenx([-0.25, 0.25], sdom_range[0], sdom_range[1], 
        color=line[0].get_color(), alpha=0.2)
legend(loc='best', ncol=4)
```

Here we can see that we've missed some of the smaller gaps that the IntervalDomain managed to pick up, which isn't really a problem in this instance. When we come to compute FFTs on this data we'll resample to have everything the same distance apart, so the effect of these gaps is likely to be minor, except at really fine frequencies...
