```python
%pylab inline

from python_boreholes import Analyst, Borehole, importers, plotting

# Borehole infomation
borehole_name = 'DET Brukunga 1'
borehole = Borehole(verbose=False)

# Add geochemistry data
labels = {
    'Ca': 'calcium',
    'S': 'sulphur',
    'Si': 'silicon',
    'Zr': 'zirconium',
    'Ti': 'titanium',
    #'Ca/Ti': 'Ca/Ti ratio (pXRF)',
    #'S/Ti': 'S/Ti ratio (pXRF)',
    #'Si/Ti': 'Si/Ti ratio (pXRF)',
    #'Zr/Ti': 'Zr/Ti ratio (pXRF)'
}
importers.add_csv(borehole, 'DETBrukunga1_pXRF.csv', domain_key='Depth', data_keys=labels.keys(), labels=labels)

# Add petrophysical data
importers.add_detcrc_xml(borehole)

# Generate an analyst instance
borehole.resample(normalize=True, detrend_method=None)
analyst = Analyst({borehole_name: borehole})
node = analyst[borehole_name]['analyst_tree']

# Do some plotting
node.plot_borehole_data()
fig, axes = node.plot_connection_matrix()
axes.set_title('Covariance between borehole traces in DETBrukunga1 profile');
fig, axes = node.plot_connection_graph(embedding='lle')
```

```python
from cwavelets import Hermitian, Meyer, WaveletProperties
from cwavelets.plotting import wt_plot

keys = ('S/Ti', 'Si/Ti', 'Ca/Ti', 'density')
signals = node.get_signal(*keys)
domain = node.get_domain()
props = WaveletProperties(order=1, frequency_decay=1/3.)

for key, signal in signals.items():
    wav = Hermitian(domain=domain, signal=signal, properties=props)
    wt_plot(wav, lambda w: w.get_transform('transform').real, signal_orientation='vertical')
    gca().set_title(key)
fig, axes = node.plot_cusum('S/Ti', 'Si/Ti', 'Ca/Ti')
```

```python
def show_array(array, axes=None):
    if axes is None:
        axes = matplotlib.pyplot.gca()
    imshow(array, interpolation='none', cmap=get_cmap('coolwarm'))
    gca().set_aspect(array.shape[1] / array.shape[0])
```

```python
import cwavelets
from python_boreholes.domaining import WaveletLabeller
from scipy.ndimage.measurements import maximum_position, minimum_position

fig = figure(figsize=(30, 5))
labeller = WaveletLabeller(node, 'Si/Ti', cwavelets.WaveletProperties(scale_range=(None, None), number_scales=50))
show_array(labeller.labelled_array)
axes = gca()
scales = labeller.wavelet.get_scales(fourier=True)
scale_array = numpy.tile(scales, (labeller.labelled_array.shape[1], 1)).T
maxima = asarray(maximum_position(scale_array, labels=labeller.labelled_array, index=labeller.labels), dtype=int)
axes.plot(maxima[:,1], maxima[:,0], 'k.')
axes.plot(maxima[:,1], maxima[:,0]+1, 'r.')
axes.set_ylim(scale_array.shape[0]-1, 0)
axes.set_xlim(scale_array.shape[1]-1, 0)
#axes.set_aspect('equal')
axes.set_xlabel('Depth (pixel #)')
axes.set_ylabel('Scale (pixel #)')

# Plot up boundaries using a linecollection
fig = figure(figsize=(12, 8))
axe = fig.gca()
colors = get_cmap('Spectral')(numpy.linspace(0, 1, len(labeller.labels)))

# Add connections
mid_domain = (labeller.min_domain + labeller.max_domain) / 2.
connect_midpoints = lambda p, c: [[mid_domain[p], labeller.max_scales[0][p]], [mid_domain[c], labeller.max_scales[0][c]]]
for index in range(len(mid_domain)):
    parent, children = (index, labeller.connections[index])
    if parent is 'root':
        continue
    segments = [connect_midpoints(parent, child) for child in children]
    color = colors[parent]
    axe.add_collection(matplotlib.collections.LineCollection(segments, linewidths=(0.5,), linestyle='solid', colors=[color]))
for label, point in enumerate(zip(labeller.max_scales[0], mid_domain)):
    text(point[1], point[0], str(label))

segments = zip(
    numpy.vstack(numpy.transpose([labeller.min_domain, labeller.max_scales[:][0]])), 
    numpy.vstack(numpy.transpose([labeller.max_domain, labeller.max_scales[:][0]])))
axe.add_collection(matplotlib.collections.LineCollection(segments, linewidths=(3,), linestyle='solid', colors=colors))

# Plot boundaries
axe.set_xlim(labeller.domain[0], labeller.domain[-1])
axe.set_yscale('log')
#axe.set_ylim(labeller.labels[0], labeller.labels[-1]);
```

```python
from scipy.ndimage import label

# Checking labelling in wavelet labeller
array = labeller.wavelet.get_transform()
domain, signal = labeller.wavelet.get_data()
pos = numpy.zeros_like(array, dtype='float') 
neg = numpy.zeros_like(array, dtype='float')
pos[array >= 0] = 1
neg[array < 0] = 1

pos_labs, npos_labs = label(pos)
neg_labs, nneg_labs = label(neg)
fig = figure(figsize=(40, 5))
labels = (pos * pos_labs - neg * (neg_labs - 1) + nneg_labs - 1).astype(int)
show_array(labels == 19)
```
