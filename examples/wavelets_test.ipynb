```python
import numpy, scipy.signal
from matplotlib.pyplot import subplot, plot
from cwavelets import Morlet, Haar, Hermitian, Paul, WaveletProperties
from cwavelets.utilities import ReSampler, detrend
from cwavelets.plotting import plot_signal, wt_plot

# Synthetic Signals
# -----------------
#
# A basic chirp signal, we include repeated time domain
# to check that cWavelets handles this gracefully using resampling methods
def crap_chirp():
    time = numpy.linspace(0, 1, 1400)
    time[:10] = numpy.linspace(0, 0.2, 10)
    time[152:392] = numpy.linspace(0.2, 1, 392 - 152)
    signal = scipy.signal.chirp(time, 1, 1, 100)
    return time, signal

# Make up a noisy synthetic signal
time = numpy.linspace(0, 1, 100)
modulate = lambda t: 50 * (t - 0.5) * (t - 0.85)
noisy_chirp_sampler = ReSampler(time, scipy.signal.chirp(time, 1, 1, 50) \
                 + numpy.random.uniform(-0.5, 0.5, len(time)) \
                 + modulate(linspace(0, 1, len(time))), order=3)
noisy_chirp = lambda n: noisy_chirp_sampler.resample(n) 
```

```python
from cwavelets.statistics import ar1_power_spectrum
from cwavelets.plotting import plot_difference

time, signal = noisy_chirp(1000)
wav_properties = WaveletProperties(fundamental_frequency=6., scaling='log', number_scales=100, pad=True)
w = Morlet(signal, domain=time, properties=wav_properties)
w.resample(1400, order=3)
w.detrend('cubic')
#trans_ax, sig_ax, spec_ax = w.spectrogram(figsize=(8,8), signal_orientation='vertical');

# Plot expected spectrum with given autocorrelation
figure(figsize=(4, 6))
axes = gca()
spectrum = numpy.sum(w.get_transform('power'),axis=1)
scales = w.get_scales(fourier=True)
nsamples = w.properties['number_samples']
variance = w.properties['signal_variance']
norm_spectrum = spectrum / (variance * nsamples)
exp_spectrum = ar1_power_spectrum(0., w.properties['sample_spacing'] * scales)

plot_difference(axes, scales, norm_spectrum, exp_spectrum, colors=('white', 'green'), orientation='vertical')
axes.set_yscale('log')
```

Donoho and Johnstone use four functions that mimic properties of empirical data in domains where wavelets might be useful, Bumps, Blocks, HeaviSine, and Doppler. 

```python
def doppler(x):
    """
    Parameters
    ----------
    x : array-like
        Domain of x is in (0,1]
 
    """
    if not np.all((x >= 0) & (x <= 1)):
        raise ValueError("Domain of doppler is x in (0,1]")
    return np.sqrt(x*(1-x))*np.sin((2.1*np.pi)/(x+.05))
 
def blocks(x):
    """ Piecewise constant function with jumps at t.
 
        Constant scalar is not present in Donoho and Johnstone.
    """
    K = lambda x : (1 + np.sign(x))/2.
    t = np.array([[.1, .13, .15, .23, .25, .4, .44, .65, .76, .78, .81]]).T
    h = np.array([[4, -5, 3, -4, 5, -4.2, 2.1, 4.3, -3.1, 2.1, -4.2]]).T
    return 3.655606 * np.sum(h*K(x-t), axis=0)
 
def bumps(x):
    """ A sum of bumps with locations t at the same places as jumps in blocks.
        The heights h and widths s vary and the individual bumps are of the
        form K(t) = 1/(1+|x|)**4
    """
    K = lambda x : (1. + np.abs(x)) ** -4.
    t = np.array([[.1, .13, .15, .23, .25, .4, .44, .65, .76, .78, .81]]).T
    h = np.array([[4, 5, 3, 4, 5, 4.2, 2.1, 4.3, 3.1, 2.1, 4.2]]).T
    w = np.array([[.005, .005, .006, .01, .01, .03, .01, .01, .005, .008, .005]]).T
    return np.sum(h*K((x-t)/w), axis=0)
 
def heavisine(x):
    """ Sinusoid of period 1 with two jumps at t = .3 and .72
    """
    return 4 * np.sin(4*np.pi*x) - np.sign(x - .3) - np.sign(.72 - x)

# Generate plots
xs = np.linspace(0, 1, 200)
nrows, ncols, index = 4, 1, 1
funcs = [blocks(xs)]#[doppler(xs), blocks(xs), bumps(xs), heavisine(xs)]
names = ['blocks'] #['doppler', 'blocks', 'bumps', 'heavisine']
#fig = plt.figure(figsize=(12, 8))
#for f, name in zip(funcs, names):
#    ax = plt.subplot(nrows, ncols, index)
#    plot_signal(ax, xs, f)
#    ax.set_title(name)
#    ax.tick_params(labelbottom=False, labelleft=False, bottom=False, top=False, left=False, right=False)
#    index += 1
#plt.tight_layout()

#wavelets = [Morlet, Haar, Haar, Haar, Morlet] 
from cwavelets import *
wav_properties = WaveletProperties(fundamental_frequency=6., scaling='log', scale_range = (0.01, 1), number_scales=100, pad=True)

wav = Haar(signal=blocks(xs), domain=xs, properties=wav_properties)
wav.spectrogram()
#for f, name, WaveletType in zip(funcs, names, wavelets):
#    wav = WaveletType(f, domain=xs, properties=wav_properties)
#    wav.detrend('cubic')
#    wav.spectrogram()
```
