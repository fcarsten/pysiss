Loading from XML
----------------

We're going to try loading from an XML schema. First, some utilities to deal with values in the XML schema

```python
from portal_parser import *

# Try and load data from pickled file first
xmlfilename = 'detcrc_xmltree.pkl'
npoints = None
try:
    with open(xmlfilename, 'rb') as fhandle:
        xmltree = pickle.load(fhandle)
except IOError:
    xmltree = get_xml_data(npoints)
    with open(xmlfilename, 'wb') as fhandle:
        pickle.dump(xmltree, fhandle)
```

Just to check that this all works, lets plot and detrend the data

```python
import cwavelets
from cwavelets.plotting import plot_signal
import numpy

# Extract data
data = extract_all(xmltree)

# Plot up data
fig = figure(figsize=(75, 6))
ax = subplot(3, 1, 1)
plot_signal(ax, data['depth'], data['p_wave_velocity'], orientation='horizontal')
ax.set_ylabel('Depth (m)')
ax.set_xlabel('p Wave Velocity')
ax = subplot(3, 1, 2)
samp = cwavelets.utilities.ReSampler(data['depth'], data['p_wave_velocity'], order=1)
data['depth'], data['p_wave_velocity'] = samp.resample(len(data['depth']))
plot_signal(ax, data['depth'], data['p_wave_velocity'], orientation='horizontal')
ax.set_ylabel('Depth (m)')
ax.set_xlabel('Interpolated\np Wave Velocity')
ax = subplot(3, 1, 3)
cwavelets.utilities.detrend(data['p_wave_velocity'], 'cubic')
plot_signal(ax, data['depth'], data['p_wave_velocity'], orientation='horizontal')
ax.set_ylabel('Depth (m)')
ax.set_xlabel('Interpolated & Detrended\np Wave Velocity')
fig.tight_layout()
```

```python
# List of keys we don't want to process
dont_process = ['id', 'diameter', 'borehole_header_id', 'depth']
dont_process.append('resistivity') # Needs denoising
keys_to_process = set(data.keys()) - set(dont_process)

# Plot up data
fig = figure(figsize=(30, 2*len(keys_to_process)))
for i, key in enumerate(keys_to_process):
    ax = subplot(len(keys_to_process), 1, i+1)
    plot_signal(ax, data['depth'], data[key], orientation='horizontal')
    ax.set_xlabel('Depth (m)')
    ax.set_ylabel(key)
fig.tight_layout()
```

Visualising wavelet transforms
------------------------------

Some functions to aid visualisation

```python
from cwavelets import *

# Make a grid of images
def make_figure_grid(nplots, ncols=3, size=6):
    nrows = nplots / ncols
    if nrows * ncols != nplots:
        nrows += 1
    fig = figure(figsize=(ncols * size, nrows * size))
    axeses = [subplot(nrows, ncols, i+1) for i in range(nplots)]
    return fig, axeses

# Generate a single image
def plot_variable(axes, variable, cmap=None):
    if cmap is None:
        cmap = get_cmap('Paired')
    axes.imshow(variable.T, cmap=cmap, interpolation='nearest')
    axes.set_aspect(variable.shape[0] / float(variable.shape[1]))
    axes.set_xlabel(r"Fourier Scale $\lambda_{i}$")
    axes.set_xticklabels([''])
    axes.set_xticks([])
    axes.set_ylabel(r"Domain $x_{i}$")
    axes.set_yticklabels([''])
    axes.set_yticks([])

# Wavelet generating function
def wavelet_from_key(key, wavelet_type):
    properties = WaveletProperties(order=2, scale_range=(0.3, 100), number_scales=100, scaling='log', pad=True)
    wav = wavelet_type(data[key], domain=data['depth'], properties=properties)
    wav.resample(len(data['depth']), order=1)
    wav.detrend('cubic')
    return wav
```

Ok, so lets pass our data in and plot it up

```python
wav = Haar(signal=data['density'], domain=data['depth'], properties=WaveletProperties(scale_range=(0.3, 100)))
wav.resample(len(data['depth']))
wav.detrend('cubic')
wav.spectrogram(variable='transform', signal_orientation='vertical')
```

```python
# List of keys we don't want to process
dont_process = ['id', 'diameter', 'borehole_header_id', 'depth']
dont_process.append('resistivity') # Needs denoising
keys_to_process = set(data.keys()) - set(dont_process)

# Generate wavelets
wavelet_type = Morlet
wavelets = [wavelet_from_key(key, wavelet_type) for key in keys_to_process]
            
# Make a couple of plots
for func_name, func in (('amplitude', numpy.abs), ('phase', numpy.angle)):
#for func_name, func in [('amplitude', (lambda x: numpy.real(x)))]:
    fig, axeses = make_figure_grid(len(keys_to_process), size=8)
    for axes, wav, key in zip(axeses, wavelets, keys_to_process):
        values = wav.get_transform()
        plot_variable(axes, func(values), cmap=get_cmap('RdYlBu_r'))
        axes.set_title(key + ' ' + func_name)
```

```python
# List of keys we don't want to process
dont_process = ['id', 'diameter', 'borehole_header_id', 'depth']
dont_process.append('resistivity') # Needs denoising
keys_to_process = set(data.keys()) - set(dont_process)

# Generate wavelets
wavelet_type = Haar
wavelets = [wavelet_from_key(key, wavelet_type) for key in keys_to_process]

fig, axeses = make_figure_grid(len(keys_to_process), size=8)
for axes, wav, key in zip(axeses, wavelets, keys_to_process):
    values = wav.get_transform()
    plot_variable(axes, numpy.real(values), cmap=get_cmap('RdYlBu_r'))
    axes.set_title(key + ' ' + func_name)
```

Ok, so lets start combining all of these things

```python
# Pairs of wavelet transforms to cross-correlate
pairs = [
    ('p_wave_velocity', 'density'),
    ('density', 'magnetic_susceptibility'),
    ('density', 'impedance')
]
wavelet_type = Haar

# Generate plots
fig, axeses = make_figure_grid(len(pairs), size=8)
for axes, pair in zip(axeses, pairs):
    wav1 = wavelet_from_key(pair[0], wavelet_type)
    wav1.resample(len(data['depth']), order=1)
    wav1.detrend('cubic')
    wav2 = wavelet_from_key(pair[1], wavelet_type)
    wav2.resample(len(data['depth']), order=1)
    wav2.detrend('cubic')
    transform = numpy.sqrt(wav1.get_transform() * wav2.get_transform().conj())
    power = numpy.real(transform * transform.conj())
    
    # Show cross-spectral power
    plot_variable(axes, power)
    axes.set_title(r'{0} $\times$ {1}'.format(*pair))
tight_layout()
```
