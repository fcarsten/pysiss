```python
from matplotlib.collections import LineCollection

def plot_interval_data(from_depths, to_depths, values, color):
    """ Plot interval data where values apply over a range specified as from-to
    """
    segments = numpy.zeros((len(from_depths), 2, 2), float)
    segments[:, 0] = numpy.asarray([from_depths, values]).T
    segments[:, 1] = numpy.asarray([to_depths, values]).T
    intervals = LineCollection(segments, linewidth=2, colors=color, linestyle='solid')
    ax = plt.axes()
    ax.add_collection(intervals)
    ax.set_xlim(from_depths.min(), to_depths.max())
    ax.set_ylim(values.min(), values.max())
    
# Plot a  bit of the data
for name, data in df.groupby('Hole_ID'):
    data = data[100:200]
    datum = data['As_ppm']
    datum = (datum - datum.mean()) / datum.std()
    colors = [get_cmap('RdPu')(i / float(0.33*len(data))) for i in range(len(data))]
    plot_interval_data(data['From'], data['To'], datum, colors)
```

```python
import functools

def mask_all_nans(*arrays):
    """ Mask all indices where any array has a NaN.

        Example usage:

            >>> mask = mask_all_nans(array1, array2)
            >>> array1[mask]            # Both of these are guarenteed 
            >>> array2[mask]            # to be nan-free

        :param *arrays: The arrays to generate masks for. They should all have the same size or a ValueError will be raised.
        :type *arrays: `numpy.ndarrays` 
        :returns: A `numpy.boolean_array` which can be used as a mask.
    """
    # Convert arrays to numpy arrays if required
    try:
        arrays = [numpy.asarray(a, dtype=numpy.float_) for a in arrays]
    except ValueError:
        raise ValueError("Arrays supplied to mask_all_nans must be able to be "
            "converted to floats!")

    # Check that everything has the same shape
    shapes = [a.shape for a in arrays]
    if any([s != shapes[0] for s in shapes]):
        raise ValueError("Arrays supplied to mask_all_nans must be the same "
            "size (arrays have shapes {0})".format(shapes))

    # Return the non-nan indices
    return numpy.logical_not(functools.reduce(numpy.logical_or, 
                                    [numpy.isnan(a) for a in arrays]))

def find_subdomains(domain, data, coeff=10, check_data=True):
    """ Find subdomains which remove large gaps in a given dataset.

        A 'large' gap is defined as any non-NaN data for which the gap between 
        this sample and the next is more than coeff * the median sample spacing.
        This function identifies these gaps and returns a set of domain intervals 
        which are 'non-gappy' by this definition.

        :returns:
            A list of tuples of the form (start_domain_i, end_domain_i)
    """
    # Optional checks (can be turned off...)
    if check_data:
        # Mask NaNs
        nan_mask = mask_all_nans(domain, data)
        _domain = domain[nan_mask]
        _data = data[nan_mask]
        
        # Ensure domain-ordering
        order_idx = numpy.argsort(_domain)
        _domain = _domain[order_idx]
        _data = _data[order_idx]
    else:
        _domain = domain
        _data = data
    
    # Generate subdomains within the domain
    spacing = numpy.diff(_domain)
    med_spacing = numpy.median(spacing)
    gap_indices = numpy.flatnonzero(spacing > coeff * med_spacing)
    
    # Form a list of data subdomains
    subdomains = [(_domain[0], _domain[gap_indices[0]])]
    for idx in range(len(gap_indices) - 1):
        start = _domain[gap_indices[idx] + 1] # We need to start _after_ the gap!
        end = _domain[gap_indices[idx + 1]] # We end with the start of the next gap
        subdomains.append((start, end))
    subdomains.append((_domain[gap_indices[-1] + 1], _domain[-1]))
    return subdomains
```
